```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(2, 3, tight_layout=True)

#折れ線グラフ
y0 = [1, 2, -5, 2]
ax[0, 0].plot(y0)

#sin関数
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
ax[0, 1].plot(x1, y1)
ax[0, 1].set_ylim(-1, 5)
ax[0, 1].set_xticks(np.arange(0, 10, 1))

#ヒストグラム
x2 = np.random .normal(15, 5, 2000)
ax[0, 2].set_xlim(-10, 40)
ax[0, 2].set_yticks([0, 200, 400])
ax[0, 2].hist(x2, bins=10)

#散布図
np.random.seed(0)
x3 = np.random.normal(3, 5, 100)
y3 = np.random.normal(3, 5, len(x3))
ax[1, 0].scatter(x3, y3)

plt.show()
```
