```python
from matplotlib import pyplot as plt
import pandas as pd
import japanize_matplotlib
```

```python
df = pd.read_csv("/content/sample_pandas_6_merged.csv") #7.4章で作成のデータフレーム
df.head()
```

```python
cnts = df["カテゴリー"].value_counts()

cnts.plot(kind = "bar")
plt.show()
```

```python
df.groupby("商品番号")["注文数"].describe()
```

```python
#papndasちゃんと使ってない

categories = df["カテゴリー"].unique().tolist()
print(categories)

category_cnts = []
for category in categories:
  category_cnts.append(len(df[df["カテゴリー"] == category]))

print(sum(category_cnts))

#棒グラフの表示
plt.bar(x = categories, height = category_cnts)
plt.show()
```
